```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
sns.countplot(df['Ship Mode'])
print("Profit from Second Class :", df[df['Ship Mode']=='Second Class']['Profit'].sum())
print("Profit from Standard Class :", df[df['Ship Mode']=='Standard Class']['Profit'].sum())
print("Profit from First Class :", df[df['Ship Mode']=='First Class']['Profit'].sum())
print("Profit from Same Day :", df[df['Ship Mode']=='Same Day']['Profit'].sum())
print("Total Profit: ", df['Profit'].sum())
```

```
sns.countplot(df['Region'])
print("Profit in East Region :", df[df['Region']=='East']['Profit'].sum())
print("Profit in West Region :", df[df['Region']=='West']['Profit'].sum())
print("Profit in Central Region :", df[df['Region']=='Central']['Profit'].sum())
print("Profit in South Region :", df[df['Region']=='South']['Profit'].sum())
print("Total Profit: ", df['Profit'].sum())
```

```
sns.countplot(df['Category'], palette="Set3")
print("Profit on Furniture: ", df[df['Category']=='Furniture']['Profit'].sum())
print("Profit on Technology: ", df[df['Category']=='Technology']['Profit'].sum())
print("Profit on Office Supplies: ", df[df['Category']=='Office Supplies']['Profit'].sum())

print("Total Profit: ", df['Profit'].sum())
plt.figure(figsize=(20,8))
values=[6026,2121,1847]
labels= ['Office Supplies','Furniture','Technology']
plt.pie(values,labels=labels,autopct='%1.1f%%',counterclock=True)
plt.title('Categories of order')
plt.show()
```

```
sns.countplot(df['Segment'], palette="Set3")
print("Profit on Consumer Segment: ", df[df['Segment']=='Consumer']['Profit'].sum())
print("Profit on Corporate Segment: ", df[df['Segment']=='Corporate']['Profit'].sum())
print("Profit on Home Office Segment: ", df[df['Segment']=='Home Office']['Profit'].sum())
print("Total Profit: ", df['Profit'].sum())
```

```
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True,cmap='RdBu')
```

```
df.hist(figsize=(12,12),bins=50)
```

```
plt.figure(figsize=(12,7))
plt.scatter(x=df['Sales'],y=df['Profit'],c='m')
plt.title("Sale VS Profit")
plt.xlabel('Total Sale')
plt.ylabel("Profit")
plt.show()
```

```

plt.figure(figsize=(12,10))
df['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
```

```
plt.figure(figsize=(15,9))
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show()
```

```
plt.figure(figsize=(12,7))
sns.countplot(x='Sub-Category',hue='Region',data=df)
plt.xticks(rotation=90)
plt.show()
```
